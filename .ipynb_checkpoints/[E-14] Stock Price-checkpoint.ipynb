# STEP0 : ARIMAÎ™®Îç∏Ïóê ÎåÄÌïú Í∞ÑÎûµÌïú ÏÑ§Î™Ö

`ARIMAÎ™®Îç∏`ÏùÄ Ï∂îÏÑ∏Ï∞®Î∂ÑÏù¥ Ï†ÅÏö©Îêú ARÍ≥º MAÏù¥ Í≤∞Ìï©Îêú Î™®Îç∏Ïù¥Îã§.
$$ \Delta^d Y_{t} = ARMA(p, q) $$

Ïù¥Îïå IÌï≠ÏùÑ Ï†úÏô∏Ìïú `ARMAÎ™®Îç∏`ÏùÄ `ARÎ™®Îç∏`Í≥º `MAÎ™®Îç∏`Ïùò Ìï©ÏúºÎ°ú ÎÇòÌÉÄÎÇúÎã§.
- `ARÎ™®Îç∏` $ Y_{t} = \phi_{1}Y_{t-1} + \phi_{2}Y_{t-2} + ... + \phi{p}Y_{t-p} + e_{t} $ 
  - Auto RegressionÏùÑ Í∞ÄÏ†ï. Îî∞ÎùºÏÑú $Y_{t}$Îäî **ÏûêÍ∏∞ÏûêÏã†**Ïùò ÏÑ†ÌòïÏ°∞Ìï©ÏúºÎ°ú Í≤∞Ï†ïÎêúÎã§Í≥† Î≥¥Îäî Î™®Îç∏.
- `MAÎ™®Îç∏` $ Y_{t} = \theta_{1}e_{t-1} + \theta_{2}e_{t-2} + ... + \theta_{q}e_{t-q} + e_{t} $
  - Moving AverageÎ•º Í∞ÄÏ†ï. Îî∞ÎùºÏÑú $Y_{t}$Îäî **Î∞±ÏÉâÏû°Ïùå**Ïùò ÏÑ†ÌòïÏ°∞Ìï©ÏúºÎ°ú Í≤∞Ï†ïÎêúÎã§Í≥† Î≥¥Îäî Î™®Îç∏.
- `ARMAÎ™®Îç∏`$ Y_{t} = \phi_{1}Y_{t-1} + \phi_{2}Y_{t-2} + ... + \phi{p}Y_{t-p} + \theta_{1}e_{t-1} + \theta_{2}e_{t-2} + ... + \theta_{q}e_{t-q} + e_{t} $ 
  - ARÍ≥º MAÎ•º ÎèôÏãúÏóê Í∞ÄÏ†ï. Îî∞ÎùºÏÑú $Y_{t}$Îäî ÏûêÍ∏∞ÏûêÏã†Ïùò ÏÑ†ÌòïÏ°∞Ìï©Ïù¥Î©∞ Î∞±ÏÉâÏû°ÏùåÏùò ÏÑ†ÌòïÏ°∞Ìï©ÏúºÎ°ú Í≤∞Ï†ïÎêúÎã§Í≥† Î≥¥Îäî Î™®Îç∏.

Îòê $Y_{t}$Ïùò Ï∂îÏÑ∏Ï∞®Î∂ÑÏùÑ Ï†ÅÏö©Ìïú ÌôïÎ•†Î≥ÄÏàò $\Delta^d Y_{t}$Ïùò `ARMAÎ™®Îç∏`ÏùÄ $Y_{t}$Ïùò `ARIMAÎ™®Îç∏`Ïù¥ ÎêúÎã§.
  - d = 1Ïùò Í≤ΩÏö∞Îäî Îã§ÏùåÍ≥º Í∞ôÏù¥ Ïì∏ Ïàò ÏûàÎã§: $Y_{t} - Y_{t-1} = \phi_{1}Y_{t-1} + ... + \phi_{p}Y_{t-p} + \theta_{1}e_{t-1} +  ... + \theta_{q}e_{t-q} + e_{t}$


ÎòêÌïú `ARMAÎ™®Îç∏`ÏùÄ Î™®Îëê $Y_{t}$Ïùò StationaryÎ•º Í∞ÄÏ†ïÌïòÎäîÎç∞, Ïù¥Îäî Ï∞®Î∂Ñ, Î°úÍ∑∏Î≥ÄÌôò Îì±ÏúºÎ°ú Ï∂îÏÑ∏ÏôÄ Í≥ÑÏ†àÏÑ±ÏùÑ Ï†úÍ±∞Ìïú ÏÉÅÌÉúÎ•º ÏùòÎØ∏ÌïúÎã§. `ARIMAÎ™®Îç∏`ÏóêÏÑúÎäî $Y_{t}$Í∞Ä Ï∂îÏÑ∏Ï∞®Î∂ÑÏùÑ ÌÜµÌï¥ StationaryÎ•º Í∞ÄÏßà Í≤ΩÏö∞ Î™®Îç∏ ÏûêÏ≤¥ÏóêÏÑú Ï∂îÏÑ∏Ï∞®Î∂ÑÏùÑ Ìï¥Ï§ÄÎã§.

> `ARIMAÎ™®Îç∏`ÏùÄ `ARÎ™®Îç∏`Í≥º `MAÎ™®Îç∏`Ïùò Í≤∞Ìï©Ïù¥Îã§. ÎòêÌïú $Y_{t}$Ïóê ÎåÄÌïú Ï∂îÏÑ∏Ï∞®Î∂ÑÏùÑ ÎÇ¥Î∂ÄÏ†ÅÏúºÎ°ú Ï†ÅÏö©Ìï¥Ï§ÄÎã§. Îî∞ÎùºÏÑú `ARIMAÎ™®Îç∏`ÏóêÏÑú Í≤∞Ï†ïÌï¥Ï£ºÏñ¥Ïïº Ìï† Î™®ÏàòÎäî p, q, dÎ°ú Í∞ÅÍ∞Å AR, MAÏùò Ï∞®ÏàòÏôÄ Ï∞®Î∂ÑÏùò ÍπäÏù¥Î°ú Í≤∞Ï†ïÎêúÎã§. 

‚Äª ÎÖºÏùòÎ•º Îî∞ÎùºÍ∞ÄÎã§Î≥¥Î©¥ ÏúÑ Î™®Îç∏Îì§Ïù¥ Î™®Îëê $Y_{t}$Ïùò Ï∂îÏÑ∏(StationaryÌïú Î∂ÄÎ∂Ñ)ÎßåÏùÑ ÏÑ§Î™ÖÌïòÍ≥† ÏûàÎäî Í≤ÉÏùÑ Ïïå Ïàò ÏûàÎã§. Ïö∞Î¶¨Í∞Ä Í∂ÅÍ∑πÏ†ÅÏúºÎ°ú ÏïåÍ≥† Ïã∂ÏùÄ Í≤ÉÏùÄ Ïù¥ ÏãúÍ≥ÑÏó¥Ïóê ÌäπÏ†ïÌïú ÏÇ¨Í±¥Ïù¥ ÏùºÏñ¥ÎÇòÏÑú Ïã§Ï†úÎ°ú ÏÉÅÏäπÌïòÎäîÏßÄ Í∞êÏÜåÌïòÎäîÏßÄ Ïùº Í≤ÉÏù¥Îã§. Í∑∏Î†áÎã§Î©¥ UnstationaryÌïú Î∂ÄÎ∂ÑÏóê ÏûàÎäî ÏûîÏ∞®Î•º ÏÑ§Î™ÖÌïòÎäî Î™®Îç∏ÏùÑ ÎßåÎì§Ïñ¥ÏïºÌïòÍ≥†, Ïù¥Í≤ÉÏùÄ `ARIMAXÎ™®Îç∏`, `SARIMAXÎ™®Îç∏` Îì±ÏóêÏÑú Ïô∏ÏÉùÎ≥ÄÏàò XÎ•º ÎèÑÏûÖÌïòÏó¨ Ìï¥Í≤∞ÌïúÎã§Í≥† ÌïúÎã§. Ìñ•ÌõÑ Ïù¥Îü¨Ìïú Î™®Îç∏Ïóê ÎåÄÌïú Í≥µÎ∂ÄÎ•º Ïù¥Ïñ¥Í∞ÄÎ≥¥Ïûê.

# STEP 3 : ÏãúÍ≥ÑÏó¥ ÏïàÏ†ïÏÑ± Î∂ÑÏÑù
- Ï†ïÏÑ±Ï†Å Í∑∏ÎûòÌîÑ Î∂ÑÏÑù
- Ï†ïÎüâÏ†Å Augmented Dicky-Fuller Test
- ÏãúÍ≥ÑÏó¥ Î∂ÑÌï¥(Time Series Decomposition)
- Residual ÏïàÏ†ïÏÑ± ÌôïÏù∏

```python
def plot_rolling_statistics(timeseries, window=12):
    
    rolmean = timeseries.rolling(window=window).mean()  # Ïù¥ÎèôÌèâÍ∑† ÏãúÍ≥ÑÏó¥
    rolstd = timeseries.rolling(window=window).std()    # Ïù¥ÎèôÌëúÏ§ÄÌé∏Ï∞® ÏãúÍ≥ÑÏó¥

     # ÏõêÎ≥∏ÏãúÍ≥ÑÏó¥, Ïù¥ÎèôÌèâÍ∑†, Ïù¥ÎèôÌëúÏ§ÄÌé∏Ï∞®Î•º plotÏúºÎ°ú ÏãúÍ∞ÅÌôîÌï¥ Î≥∏Îã§.
    orig = plt.plot(timeseries, color='blue',label='Original')    
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
```

```python
from statsmodels.tsa.stattools import adfuller

def augmented_dickey_fuller_test(timeseries):
    # statsmodels Ìå®ÌÇ§ÏßÄÏóêÏÑú Ï†úÍ≥µÌïòÎäî adfuller Î©îÏÜåÎìúÎ•º Ìò∏Ï∂úÌï©ÎãàÎã§.
    dftest = adfuller(timeseries, autolag='AIC')  
    
    # adfuller Î©îÏÜåÎìúÍ∞Ä Î¶¨ÌÑ¥Ìïú Í≤∞Í≥ºÎ•º Ï†ïÎ¶¨ÌïòÏó¨ Ï∂úÎ†•Ìï©ÎãàÎã§.
    print('Results of Dickey-Fuller Test:')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)' % key] = value
    print(dfoutput)
```

```python
# Ï†ïÏÑ±Ï†Å Í∑∏ÎûòÌîÑ Î∂ÑÏÑù
plot_rolling_statistics(ts_log, window=12)
```

```python
#Ï†ïÎüâÏ†Å Augmented Dicky-Fuller Test
augmented_dickey_fuller_test(ts_log)
```

 - dickey fuller testÏùò Í∑ÄÎ¨¥Í∞ÄÏÑ§ÏùÑ Í∏∞Í∞ÅÌïòÏßÄ Î™ªÌï¥ ÎπÑÏ†ïÏÉÅÏÑ±ÏùÑ Îî∞Î•¥Îäî Îç∞Ïù¥ÌÑ∞ÎùºÍ≥† Î≥º Ïàò ÏûàÎã§. Îî∞ÎùºÏÑú ÏãúÍ≥ÑÏó¥ Î∂ÑÌï¥Î•º ÏãúÎèÑÌïúÎã§.

```python
#ÏãúÍ≥ÑÏó¥ Î∂ÑÌï¥ (Time Series Decomposition)
decomposition = seasonal_decompose(ts_log, model='multiplicative', period = 30) 

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
plt.plot(ts_log, label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
```

```python
residual.dropna(inplace=True)
augmented_dickey_fuller_test(residual)
```

- ÏûîÏ∞®Îäî dickey fuller testÏùò Í∑ÄÎ¨¥Í∞ÄÏÑ§ÏùÑ Í∏∞Í∞ÅÌïòÍ≥† Ï†ïÏÉÅÏÑ±ÏùÑ ÎßåÏ°±ÌïúÎã§. Îî∞ÎùºÏÑú ÏûîÏ∞®Îäî Ï†ïÏÉÅÏÑ±ÏùÑ Îî∞Î•¥Îäî Î∞±ÏÉâÏû°ÏùåÏù¥Îã§!

# STEP 6 : ARIMA Î™®Îç∏ ÌõàÎ†®Í≥º ÌÖåÏä§Ìä∏

- Ïó¨Îü¨ Î™®Ìòï Ï§ë Log LikelihoodÏùò ÏµúÎåÄÍ∞íÏùÑ Í∞ñÎäî ÌååÎùºÎØ∏ÌÑ∞Î•º Ï∞æÏïÑ Í∏∞Ï§ÄÏúºÎ°ú Ìï¥Î≥¥Ïûê.

```python
# ARIMA(p,d,q) Î™®Îç∏ÎßÅ
result = []
for p, d, q in product(range(1,2), range(1,3), range(1,10)):
    if (p == 0 & q == 0):
        continue
    model = sm.tsa.ARIMA(train_data, (p, d, q)).fit()
    try:
        result.append({"p": p, "d": d, "q": q, "LLF": model.llf, "AIC": model.aic, "BIC": model.bic})
    except:
        pass

#Î™®Ìòï ÏµúÏ†ÅÎ™®Ïàò ÏÑ†ÌÉù
result = pd.DataFrame(result)
display(result)


opt_ar = result.iloc[np.argmax(result['LLF']), 0]
opt_i = result.iloc[np.argmax(result['LLF']), 1]
opt_ma = result.iloc[np.argmax(result['LLF']), 2]
```

```python
plt.subplot(121)
result['LLF'][result['d'] == 1].plot()
plt.subplot(122)
result['LLF'][result['d'] == 2].plot()
```

- likelihoodÍ∞íÏùÑ Î≥¥Î©¥ 
  - dÎäî 1Ïù∏ Í≤ΩÏö∞Í∞Ä Ï†ÑÏ≤¥Ï†ÅÏúºÎ°ú ÌÅ¨Îã§.
  - qÎäî Ï∞®ÏàòÏùò Ï¶ùÍ∞ÄÏóê Îî∞Îùº ÌÅ∞ Ï∞®Ïù¥Í∞Ä ÏóÜÏùå. Îî∞ÎùºÏÑú Ï≤òÏùå Ï¶ùÍ∞ÄÌïòÎäî Î∂ÄÎ∂ÑÏù∏ 4 Ï†ïÎèÑÎ•º Í∏∞Ï§ÄÏúºÎ°ú ÌïúÎã§.

```python
# ARMA Î™®Îç∏ÎßÅ
fitted_m = sm.tsa.ARIMA(train_data, (1,1, 1)).fit(disp=-1)
display(fitted_m.summary())
```

```python
# ARMA Î™®Îç∏ÎßÅ
fitted_m = sm.tsa.ARIMA(train_data, (1,1, 4)).fit(disp=-1)
display(fitted_m.summary())
```

- (1, 1, 4) Î™®Îç∏ÏóêÏÑú p value Î•º Î≥¥Î©¥ ARÏùò Ï≤´Ïß∏Ìï≠Í≥º MAÏùò Ï≤´Ïß∏Ìï≠Ïù¥ Ïú†ÏùòÎØ∏Ìï®ÏùÑ Ïïå Ïàò ÏûàÎã§.

```python
fitted_m.plot_predict()
```

```python
# Forecast : Í≤∞Í≥ºÍ∞Ä fcÏóê Îã¥ÍπÅÎãàÎã§. 
fc, se, conf = fitted_m.forecast(len(test_data), alpha=0.05)  # 95% conf

# Make as pandas series
fc_series = pd.Series(fc, index=test_data.index)   # ÏòàÏ∏°Í≤∞Í≥º
lower_series = pd.Series(conf[:, 0], index=test_data.index)  # ÏòàÏ∏°Í≤∞Í≥ºÏùò ÌïòÌïú Î∞îÏö¥Îìú
upper_series = pd.Series(conf[:, 1], index=test_data.index)  # ÏòàÏ∏°Í≤∞Í≥ºÏùò ÏÉÅÌïú Î∞îÏö¥Îìú

# Plot
plt.figure(figsize=(10,5), dpi=100)
plt.plot(train_data, label='training')
plt.plot(test_data, c='b', label='actual price')
plt.plot(fc_series, c='r',label='predicted price')
plt.fill_between(lower_series.index, lower_series, upper_series, color='k', alpha=.10)
plt.legend()
plt.show()
```

```python
mse = mean_squared_error(np.exp(test_data), np.exp(fc))
print('MSE: ', mse)

mae = mean_absolute_error(np.exp(test_data), np.exp(fc))
print('MAE: ', mae)

rmse = math.sqrt(mean_squared_error(np.exp(test_data), np.exp(fc)))
print('RMSE: ', rmse)

mape = np.mean(np.abs(np.exp(fc) - np.exp(test_data))/np.abs(np.exp(test_data)))
print('MAPE: {:.2f}%'.format(mape*100))
```
